**Бутстрап для медиан разностей**

```python
import numpy as np

matrix1 = f1_srv_ord
matrix2 = f2_srv_ord

median_points = []
ci_lows = []
ci_highs = []
param = 'rows'

# Предполагается, что matrix1 и matrix2 — это pandas DataFrame.
for i in range(matrix1.shape[0]):
    if param == 'rows':
        x = matrix1.iloc[i, :].values
        y = matrix2.iloc[i, :].values
    elif param == 'cols':
        x = matrix1.iloc[:, i].values
        y = matrix2.iloc[:, i].values
    else:
        flat1 = matrix1.values.flatten()
        flat2 = matrix2.values.flatten()

    
    
    
    n_bootstraps = 10000  # количество бутстрэп-выборок
    medians = []
    
    # Считаем разности внутри каждой пары (важно для парных данных!)
    diff = x - y
    
    rng = np.random.default_rng(seed=42)  # для воспроизводимости
    
    for _ in range(n_bootstraps):
        # Берём случайную выборку из diff с возвращением (размер тот же, что у diff)
        sample = rng.choice(diff, size=len(diff), replace=True)
        medians.append(np.median(sample))
    
    # Центральная оценка и 95% доверительный интервал
    median_point = np.median(medians)
    ci_low, ci_high = np.percentile(medians, [2.5, 97.5])
    median_points.append(median_point)

    print(f'Срез № {i+1}')
    print(f'Оценка медианы разностей: {median_point:.4f}')
    
    print(f'95% доверительный интервал: [{ci_low:.4f}; {ci_high:.4f}]')
    if ci_low <= 0 and ci_high >= 0:
        print('различие не подтверждено')
    print('-----------')


indices = range(1, matrix1.shape[0]+1)

plt.figure(figsize=(8, 5))
plt.plot(indices, median_points, marker='o', label='Оценка медианы разностей')
plt.fill_between(indices, ci_lows, ci_highs, color='b', alpha=0.2, label='95% ДИ')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.xlabel('Номер среза (i)')
plt.ylabel('Медиана разности')
plt.title('Оценка медианы разностей и доверительные интервалы по срезам')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('medians_by_slice.png', dpi=300)  # Сохраняет график в PNG с хорошим качеством
```
